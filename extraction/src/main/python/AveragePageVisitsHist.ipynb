```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pylab as P
from scipy.spatial import Voronoi, voronoi_plot_2d
```

```python
def read_data(file_name, project, value, is_float, sep=' '):
    data = []
    for i in range(1,9,1):
        print str(float(i)/8*100) + "% read"
        f = open(file_name + str(i), 'r')
        if is_float:
            data += [float(line.split(sep)[value]) for line in f if line.split(sep)[0] == project]
        else:
            data += [line.split(sep)[value] for line in f if line.split(sep)[0] == project]
    print "converting... almost finished"
    return np.array(data)
```

```python
data = read_data("../../test/resources/", "de", 2, True)
```

```python
len(data)
```

```python
filter_large = data > 1000
filter_small = data < 25

filter_medium = np.logical_and(data > 5, data < 100)
```

```python
plt.hist(data[filter_small], 50)
plt.show()
```

```python
plt.hist(data[filter_medium], 95)
plt.show()
```

```python
plt.hist(data[filter_large], 100)
plt.show()
```
